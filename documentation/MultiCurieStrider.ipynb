# MultiCurie AC Strider

This version of Strider finds and ranks all paths that fulfill a MultiCurie query such as what `chemicals` regulates a group of `genes`. 

<code> query_graph: { 

            nodes: { node_0: {categories: Gene, ids:[Gene1, Gene2, Gene3, ...], is_set=True}, 
            
                    node_1: { categories: [ChemicalEntity] } }, 

            edges: { edge_x: { source: node_0, target: node_1, predicates: [..] } 
            
        }
</code>

- [Output Formating](https://docs.google.com/presentation/d/12hb6UCdOrHDs9fI-70FpvQojf2Q-mbyDCnrmVNeTFxM/edit#slide=id.g1f1e07b1889_0_49)
- [Concept](https://docs.google.com/presentation/d/1kSFSFD-FzQGzMlkbIuP1jps0M-3UjBFl/edit#slide=id.p1)

```python
import requests
import json

url = "https://answercoalesce-dev.apps.renci.org/1.4/query/"
```

```python
Samplequery = {
    "message": {
        "query_graph": {
            "nodes": {
                "chemical": {
                    "categories": [
                        "biolink:ChemicalEntity"
                    ],
                    "is_set": False,
                    "constraints": []
                },
                "gene": {
                    "ids": ["NCBIGene:3952", "NCBIGene:9370", "NCBIGene:5468"],
                    "categories": [
                        "biolink:Gene"
                    ],
                    "is_set": True,
                    "constraints": []
                }
            },
            "edges": {
                "e00": {
                    "subject": "chemical",
                    "object": "gene",
                    "predicates": [
                        "biolink:affects"
                    ],
                    "knowledge_type": "inferred",
                    "attribute_constraints": [],
                    "qualifier_constraints": [
                        {
                            "qualifier_set": [
                                {
                                    "qualifier_type_id": "biolink:object_aspect_qualifier",
                                    "qualifier_value": "activity_or_abundance"
                                },
                                {
                                    "qualifier_type_id": "biolink:object_direction_qualifier",
                                    "qualifier_value": "increased"
                                }
                            ]
                        }
                    ]
                }
            }
        }
  }
}

response = requests.post(url,json=Samplequery)
print(response.status_code)
response.json()
```
